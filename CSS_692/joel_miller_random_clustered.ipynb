```python
import networkx as nx
import random
import matplotlib.pyplot as plt
```

```python
n = 100
alpha = 2
n_tk = []
for i in range(0,alpha+1):
    n_tk.append([alpha-i,i*2])

t = []
k = []

for v in range(n):
    i = v+1
    param = random.choice(n_tk)
    for x in range(param[0]):
        t.append(i)
    for x in range(param[1]):
        k.append(i)
    
random.shuffle(t)
random.shuffle(k)

G = nx.Graph()

for i in range(len(t)//3):
    x = i*3
    if t[x] != t[x+1] != t[x+2]:
        G.add_edge(t[x],t[x+1])
        G.add_edge(t[x],t[x+2])
        G.add_edge(t[x+1],t[x+2])

for i in range(len(k)//2):
    x = i*2
    if k[x] != k[x+1]:
        G.add_edge(k[x],k[x+1])
        
nx.draw_networkx(G)
plt.show()
plt.hist(list(nx.clustering(G).values()))
plt.show()
```
