```python
import networkx as nx
import random
```

```python
H = nx.Graph() #H is the symbol for bipartite in our networks
V_e = [1,2,3,4] # entity notes class (greek letters)
V_a = ['a','b','c','d'] # actor nodes class (latin letters)
# Each edge has a vertex in e and a
# Actors should be columns, entities should be rows in the ADJ Matrix
P = 0.6
H.add_nodes_from(V_e,bipartite=0)
H.add_nodes_from(V_a,bipartite=1)

for i in V_e:
    for j in V_a:
        r = random.random()
        print(i,j,r)
        if random.random() < P:
            H.add_edge(i,j)
H.edges()

nx.draw_networkx(H)
```
