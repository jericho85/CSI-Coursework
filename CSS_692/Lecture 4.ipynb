given that Degree is an integer, we can use python to create a histogram with Dicts.

```python
import random

k  = []
for i in range(0,100):
    k.append(random.randint(0,10))

H = {}
for i in k:
    H[i] = H.get(i,0)+1
# Awesomoe way to do a dictionary

for k,v in sorted(H.items()):
    print(k,v)
```

```python
import matplotlib.pyplot as plt
import math
```

```python
x = []
y = []
z = []
t = []
for i in range(0,10000):
    x.append(i/1000)
    y.append(math.sin(i/1000))
    z.append(math.cos(i/1000))
    t.append(math.tan(i/1000))
```

```python
plt.plot(x,y)
plt.plot(x,z)
plt.plot(x,t)
plt.ylim(-1.2,1.2)
plt.show()
```

```python
e = []
f = [] 
for i in x:
    e.append(i**2)
    f.append((i**0.5)*(31.5))
plt.plot(x,e)
plt.plot(x,f)
plt.ylim(-10,110)
plt.show()
```
