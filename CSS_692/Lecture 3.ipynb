```python
import networkx as nx
import matplotlib.pyplot  as plt
```

```python
#This is a Watts-Strogatz network!
G  = nx.Graph()
edges_input = [[0,2],\
               [0,5],\
               [1,6],\
               [1,3],\
               [2,4],\
               [3,5],\
               [4,6],\
               [0,1],\
               [2,1],\
               [2,3],\
               [3,4],\
               [4,5],\
               [5,6],\
               [0,6]
              ]
for i in edges_input:
    G.add_edge(i[0],i[1])
```

```python
nx.draw_networkx(G)
```

Number  of hops in longest shortest path:
$$\frac{(\frac{n}{2})}{(\frac{k}{2})} = \frac{n}{k}$$
The example above is growing linear in paths, not the log of the growth rate. <br>
This is called a <b> Large World Network </b>, because < s > is proportional to n to some power. <br>
$< s >  \alpha n^{\gamma}$  where  alpha is some relationship.<br>

Playing with Watts Stogatz networks

```python
def build_WS(n,k):
    G  = nx.Graph()
    for i in  range(0,n-1):
        for c in range(1,int((k/2)+1)):
            j=i+c
            if j>n-1:
                j=j-n
            G.add_edge(i,j)
    return(G)

G  = build_WS(10,2)
nx.draw_networkx(G)
```
