Centralities -or- Centrality Measures:<br>
Families of measures to assign importance / relevance to individual nodes (links), generally capturinig the relation between the ndoe and the network.<br>
1) Betweenness Centrality<br>
2) Eigenvector Centrality<br>

Betweenness Centrality ->  Bare betweenness: 

1) A set of paths from every (origin) node  to every (destination) node<br>
2) Betweenness: node i has betweenness $b_i$ <br>
$\;\;\;\;\;\;$ where $b_i$ is the # of paths i is in involved in

The common choice for types of path is shortest  path in the network.<br>
${n  \choose k}$ shortest paths exist in a network<br>
As homogeneity increases in a network, the importance of the rule defining paths increases<br>

Always use  a network you can verify by hand to verify betweenness in a new library. Examples of graphs useful for this are shown below

```python
import networkx as nx
G=nx.Graph()
G.add_node(1)
G.add_node(2)
G.add_node(3)
G.add_node(4)
G.add_edge(1,2)
G.add_edge(1,4)
G.add_edge(2,4)
G.add_edge(2,3)
G.add_edge(3,4)
G.add_edge(3,5)
G.add_edge(1,6)
nx.draw_networkx(G)
```

```python
G=nx.Graph()
for i in range(5):
    G.add_edge(i+1,i+2)
G.add_edge(1,6)
G.add_edge(3,6)
G.add_edge(5,7)
nx.draw_networkx(G)
```

Shortest paths are determined by using a version of Breadth-First-Search (burning search in physics).<br>
Start with A origin, and find all possible  destinations.<br>These are then called Shell L = 1.<br>  All new shortest paths go to the next set of nodes, from all nodes in shell L = 1.<br> This becomes shell L = 2.

From  graph above you can see that we have to deal with multiplicity, or multiple ways to get to a node.


```python
nx.betweenness_centrality(G)
```

$$b_i^{(o)} = \upsilon_i^{(o)} \sum_{j_{neighbor-successor: i}} \frac{\upsilon_i^{(o)}}{\upsilon_j^{(o)}}b_j^{(o)}$$

$$b_i = \frac{1}{2}\sum_{o=1}^{n}b_i^{(o)}$$
