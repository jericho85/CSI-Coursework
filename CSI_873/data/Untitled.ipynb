```python
import matplotlib.pyplot as plt
data = [0,1,0,0,1,1,0,1]

x=[]
y=[]

for i in range(len(data)):
    data1 = data[:i]
    data2 = data[i:]
    data1_count = 0
    data2_count = 0
    for j in data1: 
        if j==0:
            data1_count+=1
    for j in data2:
        if j==0:
            data2_count+=1
    if len(data1)>0:
        data1_count = data1_count/len(data1)
        x.append(data1_count)
    else: x.append(0)
    if len(data2) >0:
        data2_count = data2_count/len(data2)
        y.append(data2_count)
    else: y.append(0)

plt.plot(x,y)
plt.show()
```
